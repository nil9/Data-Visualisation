```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import math
from mpl_toolkits.mplot3d import Axes3D


# readout the header
with open("field2.irreg.txt", "r") as f:
        rows = f.readlines()[6:]


count = 0

X = []
Y = []
U = []
V = []
R = []



#readout Data
for line in rows:
    s = line.split(" ")

    X.append(float(s[0]))
    Y.append(float(s[1]))
    U.append(float(s[3]))
    V.append(float(s[4]))
    # resultant list
    R.append(float(s[3])+float(s[4]))
    
    count+=1

f.close()

fig, ax = plt.subplots(figsize=(10,7))
ax.quiver(X,Y,U,V,R,scale=4,pivot='tip')
ax.set_title('Water Direction')

cax = fig.add_axes([0.75,0.075,0.07,0.85])
norm = mpl.colors.Normalize(vmin = 0, vmax = 1)
color_base = mpl.colorbar.ColorbarBase(cax,norm = norm, orientation = 'vertical')
norm = mpl.colors.Normalize(vmin = 0, vmax = 1)
color_base.set_ticks([0.2,0.6,0.9])
color_base.set_ticklabels(['Low', 'Medium', 'High']) 

color_base.set_label('Velocity')


plt.show()
```
