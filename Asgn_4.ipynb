```python
from pandas.plotting import scatter_matrix

import pandas as pd
import numpy as np
import csv
import plotly.graph_objs as go
from plotly.offline import iplot
import matplotlib.pyplot as plt
import heapq 


##################
##Task 1##########
##################






data = {}
header = {}
count = 0
flist = []

with open('DataWeierstrass.csv') as csv_file:
    csv_data = csv.reader(csv_file, delimiter=',')

    for row in csv_data:
        
        line = str(row).replace("[", "").replace("]", "").replace("'","").split(";")
        #print(d)
        dataRow = []

        if count==0:
            for i in range(0,len(line)):
                header[i] = line[i]
            count+= 1
        else:

            for i in range(0,len(line)):
                dataRow.append(float(line[i].replace("prof", "").replace("lecture", "")))
                
            flist.append(dataRow)


    df = pd.DataFrame(flist,columns = [header[0], header[1], header[2], header[3], header[4], header[5], header[6]])


    scatter_matrix(df, diagonal = 'hist')
    
    ##################
    ##Task 2##########
    ##################



data = [
    go.Parcoords(
            dimensions = list([
            dict(range = [0,44],
                 label = 'Professor', values = df['professor']),
            dict(range=[0, 101],
                 label='Lecture', values=df['lecture']),
            dict(range = [0,326],
                 label = 'Participants', values = df['participants']),
            dict(range=[1, 2.7],
                 label='Professional Expertise', values=df['professional expertise']),
            dict(range = [1,3.6],
                 label = 'Motivation', values = df['motivation']),
            dict(range=[1, 3.47],
                 label='Clear Presentation', values=df['professional expertise']),
            dict(range = [1,4.06],
                 label = 'Overall Impression', values = df['overall impression'])
        ])
    )
]

iplot(data)

csv_file.close()
plt.show()



```
